## Import libraries

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

## Extracting and Parsing Data from URL

```python
URL = 'http://www.millenniumcricketleague.com/Home/PreviousSeason.aspx'
page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')
table = soup.find('table')
```

```python
table_data= table.find_all('tr')
x=table_data[0].find_all("tr")
```

```python
headings=[]
for i in x[0].find_all("th"):
    headings.append(i.text.replace('\n', ' ').strip())
```

```python
headings
```

```python
headings = [ elem for elem in headings if elem != ""] 
```

```python
for h3 in table_data[0].find_all("h3"):
    if h3.text.replace('\n', ' ').strip()== "Forty 40":
        Forty_40_data=table_data[0].find(id= "ctl00_MainContent_rptSeason_ctl00_grdPreviousSeasons")

    elif h3.text.replace('\n', ' ').strip()== "Twenty 20":
        Twenty_20_data=table_data[0].find(id= "ctl00_MainContent_rptSeason_ctl01_grdPreviousSeasons")
```

```python
Forty_40_table_data=[]
Twenty_20_table_data =[]

for tr in Forty_40_data.find_all("tr"):
    t_row = {}
    for td, th in zip(tr.find_all("td"), headings):
        t_row[th] = td.text.strip()

    Forty_40_table_data.append(t_row)
df_Forty_40_table_data= pd.DataFrame(Forty_40_table_data)

for tr in Twenty_20_data.find_all("tr"):
    t_row = {}
    for td, th in zip(tr.find_all("td"), headings):
        t_row[th] = td.text.strip()

    Twenty_20_table_data.append(t_row)
df_Twenty_20_table_data= pd.DataFrame(Twenty_20_table_data)
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', -1)
```

# Data Cleaning

### Forty 40

```python
df_Forty_40_table_data=df_Forty_40_table_data.drop(df_Forty_40_table_data.index[0])
df_Forty_40_table_data=df_Forty_40_table_data.sort_values(by=['Year'])
df_Forty_40_table_data=df_Forty_40_table_data.reset_index(drop=True)
```

```python
df_Forty_40_table_data
```

### Twenty 20

```python
df_Twenty_20_table_data=df_Twenty_20_table_data.drop(df_Twenty_20_table_data.index[0])
df_Twenty_20_table_data=df_Twenty_20_table_data.reset_index(drop=True)
```

```python
df_Twenty_20_table_data
```

```python
writer = pd.ExcelWriter('Scraped_Data_previous_season.xlsx')
```

```python
df_Forty_40_table_data.to_excel(writer, sheet_name='Sheet1', index=False)
df_Twenty_20_table_data.to_excel(writer, sheet_name='Sheet2', index=False)
```
